```python
import pandas as pd
import matplotlib.pyplot as plt

model_df = pd.read_csv(
    "/Users/aaronrao/Desktop/projects/UIDAI_Aadhaar_Data_Insights/outputs/tables/aadhaar_service_stress_model_output.csv"
)

model_df.head()
```

```python
import matplotlib.pyplot as plt

state = "Karnataka"      # change during demo
district = "Bangalore Urban"

region_df = model_df[
    (model_df["state"] == state) &
    (model_df["district"] == district)
]

if region_df.empty:
    print("No data available for the selected region.")
else:
    row = region_df.iloc[0]

    # Print summary
    print(f"State: {row['state']}")
    print(f"District: {row['district']}")
    print(f"Total Enrolments: {int(row['total_enrolments']):,}")
    print(f"Biometric Updates: {int(row['total_biometric_updates']):,}")
    print(f"Demographic Updates: {int(row['total_demographic_updates']):,}")
    print(f"Service Stress Score: {row['service_stress_score']:.2f}")
    print(f"Category: {row['stress_category']}")

    # Plot gauge-style indicator
    plt.figure(figsize=(6, 1.5))
    plt.barh(
        ["Service Stress"],
        [row["service_stress_score"]],
        color="red"
    )

    plt.xlim(0, model_df["service_stress_score"].max())
    plt.title("Aadhaar Service Stress Indicator")
    plt.show()
```

```python
model_df["state"].unique()[:20]
```

```python
# Remove rows where state is numeric or too short
model_df = model_df[
    model_df["state"].str.contains(r"[A-Za-z]", regex=True)
].copy()
```

```python
state_mapping = {
    "Andaman And Nicobar Islands": "Andaman & Nicobar Islands",
    "Dadra And Nagar Haveli": "Dadra & Nagar Haveli",
    "Daman And Diu": "Daman & Diu",
    "Dadra And Nagar Haveli And Daman And Diu": "Dadra & Nagar Haveli And Daman & Diu",
    "Jammu & Kashmir": "Jammu & Kashmir"
}

model_df["state"] = model_df["state"].replace(state_mapping)
```

```python
sorted(model_df["state"].unique())
```

```python
# Pick top stress district (always valid)
row = model_df.sort_values(
    by="service_stress_score",
    ascending=False
).iloc[0]

print(f"State: {row['state']}")
print(f"District: {row['district']}")
print(f"Service Stress Score: {row['service_stress_score']:.2f}")
print(f"Category: {row['stress_category']}")
```
